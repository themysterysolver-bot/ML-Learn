```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('LINEAR_REG.xlsx')
```

```python
print(df)
```

```python
xm,ym=df['x'].mean(),df['y'].mean()
print(xm,ym)
```

```python
df['x-xm']=df['x']-xm
df['y-ym']=df['y']-ym
print(df)
```

```python
df['(x-xm)*(y-ym)']=df['x-xm']*df['y-ym']
print(df)
```

```python
df['(x-xm)**2']=df['x-xm']**2
print(df)
```

```python
b=df['(x-xm)*(y-ym)'].sum()/df['(x-xm)**2'].sum()
print(b)
```

```python
a=ym-b*xm
print(a)
```

```python
def linear_model(x):
    return b*x+a
print(linear_model(3))
```

```python
df.to_excel('LINEAR_REG_RESULT.xlsx')
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['x'],df['y'],marker='*',label=f'y={a}+{b}*x')
xc=np.linspace(1,9,10)
#print(xc)
yc=linear_model(xc)
#print(yc)
plt.plot(xc,yc,color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
